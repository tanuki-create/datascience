## ローカル閾値

```python
# 適応的閾値処理の詳細img = np.zeros((200, 200), dtype=np.uint8)img[50:150, 50:150] = 200for i in range(200):    img[:, i] = np.clip(img[:, i].astype(np.int16) + i // 4, 0, 255).astype(np.uint8)adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)plt.figure(figsize=(10, 5))plt.subplot(1, 2, 1)plt.imshow(img, cmap='gray')plt.title('元の画像')plt.axis('off')plt.subplot(1, 2, 2)plt.imshow(adaptive, cmap='gray')plt.title('適応的閾値')plt.axis('off')plt.tight_layout()plt.show()
```
