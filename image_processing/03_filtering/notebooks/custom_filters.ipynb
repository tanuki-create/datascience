```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
from skimage import filters, feature

plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams["axes.unicode_minus"] = False
np.random.seed(42)
print("ライブラリのインポートが完了しました。")
```

## 畳み込み演算

```python
# カスタムカーネルの作成kernel = np.array([    [-1, -1, -1],    [-1,  8, -1],    [-1, -1, -1]], dtype=np.float32)# 畳み込み演算の適用img = np.random.randint(0, 256, (200, 200), dtype=np.uint8)filtered = cv2.filter2D(img, -1, kernel)plt.figure(figsize=(10, 5))plt.subplot(1, 2, 1)plt.imshow(img, cmap='gray')plt.title('元の画像')plt.axis('off')plt.subplot(1, 2, 2)plt.imshow(filtered, cmap='gray')plt.title('カスタムフィルタ適用後')plt.axis('off')plt.tight_layout()plt.show()
```

## カスタムカーネル

```python
# 様々なカスタムカーネルimg = np.random.randint(0, 256, (200, 200), dtype=np.uint8)# エッジ検出カーネルedge_kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]], dtype=np.float32)# シャープ化カーネルsharpen_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]], dtype=np.float32)# エンボスカーネルemboss_kernel = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]], dtype=np.float32)edge = cv2.filter2D(img, -1, edge_kernel)sharpen = cv2.filter2D(img, -1, sharpen_kernel)emboss = cv2.filter2D(img, -1, emboss_kernel)fig, axes = plt.subplots(2, 2, figsize=(10, 10))axes[0,0].imshow(img, cmap='gray')axes[0,0].set_title('元の画像')axes[0,0].axis('off')axes[0,1].imshow(edge, cmap='gray')axes[0,1].set_title('エッジ検出')axes[0,1].axis('off')axes[1,0].imshow(sharpen, cmap='gray')axes[1,0].set_title('シャープ化')axes[1,0].axis('off')axes[1,1].imshow(emboss, cmap='gray')axes[1,1].set_title('エンボス')axes[1,1].axis('off')plt.tight_layout()plt.show()
```
